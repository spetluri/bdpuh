```python
from kafka import KafkaConsumer
import io
import avro.schema
import avro.io
from avro.datafile import DataFileReader, DataFileWriter
from avro.io import DatumReader, DatumWriter
import happybase
```

```python
consumer = KafkaConsumer('movie_details')
```

```python
schema_path = "movie_info.avsc"
schema = avro.schema.parse(open(schema_path).read())


connection = happybase.Connection('localhost')
moviedetailstable = connection.table('moviedetails')

for msg in consumer:
    value = bytearray(msg.value)
    bytesReader = io.BytesIO(value)
    decoder = avro.io.BinaryDecoder(bytesReader)
    reader = avro.io.DatumReader(schema)
    json = reader.read(decoder)
    print(json)
    moviedetailstable.put(str(json['movieId']),{"details:moviedetails":msg.value})
```
