```python
import csv
import io
import avro.schema
import avro.io
from avro.datafile import DataFileReader, DataFileWriter
from avro.io import DatumReader, DatumWriter
from kafka import KafkaProducer, KafkaConsumer
```

```python
producer = KafkaProducer(bootstrap_servers='localhost:9092')

def convertToBoolean(num):
    if num == '0':
        return False
    if num == '1':
        return True

def createMovieInfoRecord(row):
    return {
        
        "movieId":int(row[0]),
        "movieTitle":row[1],
        "releaseDate":row[2],
        "videoReleaseDate":row[3],
        "imdbUrl":row[4],
        "movieGenres":{
            "unknown":convertToBoolean(row[5]),
            "action":convertToBoolean(row[6]),
            "adventure":convertToBoolean(row[7]),
            "animation":convertToBoolean(row[8]),
            "childrens":convertToBoolean(row[9]),
            "comedy":convertToBoolean(row[10]),
            "crime":convertToBoolean(row[11]),
            "documentary":convertToBoolean(row[12]),
            "drama":convertToBoolean(row[13]),
            "fantasy":convertToBoolean(row[14]),
            "filmNoir":convertToBoolean(row[15]),
            "horror":convertToBoolean(row[16]),
            "musical":convertToBoolean(row[17]),
            "mystery":convertToBoolean(row[18]),
            "romance":convertToBoolean(row[19]),
            "sciFi":convertToBoolean(row[20]),
            "thriller":convertToBoolean(row[21]),
            "war":convertToBoolean(row[22]),
            "western":convertToBoolean(row[23])
        }
    }


schema_path = "movie_info.avsc"
schema = avro.schema.parse(open(schema_path,"rb").read())

with open("./ml-100k/u.item", newline='', encoding = "ISO-8859-1") as movieInfoFile:
    reader = csv.reader(movieInfoFile,delimiter='|')
    for row in reader:
        writer = DatumWriter(schema)
        bytes_writer = io.BytesIO()
        encoder = avro.io.BinaryEncoder(bytes_writer)
        writer.write(createMovieInfoRecord(row),encoder)
        raw_bytes = bytes_writer.getvalue()
        producer.send("movie_details",raw_bytes)
        
```

```python
def create_user_rating_record(row):
    return {
    
        'userId':int(row[0]),
        'movieId':int(row[1]),
        'movieRating':int(row[2]),
        'timestamp':int(row[3])
        }


schema_path = "user_rating.avsc"
schema = avro.schema.parse(open(schema_path).read())





count = 100
with open("./ml-100k/u.data", newline='') as csvfile:
    reader = csv.reader(csvfile,delimiter='\t')
    for row in reader:
        
        if count == 0:
            break
    
        print(count)
        
        writer = DatumWriter(schema)
        bytes_writer = io.BytesIO()
        encoder = avro.io.BinaryEncoder(bytes_writer)
        writer.write(create_user_rating_record(row),encoder)
        raw_bytes = bytes_writer.getvalue()
        producer.send("movie_ratings",raw_bytes)
        count-=1
        
        
        
```
